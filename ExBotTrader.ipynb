##2.1. Obtendo preços da BINANCE através da API

```
#Pegando preços intervalo de 1 dia

BTCUSDT = client.get_klines(symbol='BTCUSDT', interval=Client.KLINE_INTERVAL_1DAY)
```

```
BTCUSDT
```

```
#transformando o json

import json
import pandas as pd

with open('btc_df.json', 'w') as e:
    json.dump(BTCUSDT, e)
```

```
for line in BTCUSDT:
    del line[5:]

```

```
btc_df = pd.DataFrame(BTCUSDT, columns=['date', 'open', 'high', 'low', 'close'])
btc_df.set_index('date', inplace=True)
btc_df.index = pd.to_datetime(btc_df.index, unit='ms')
print(btc_df.head())
```

```
btc_df['close'] = pd.to_numeric(btc_df['close'])
```

```
#Preços de Fechamento

btc_df['close'].plot();
```

```
#media_movel

import matplotlib.pyplot as plt

mm_30 = btc_df.close.rolling(30).mean();
```

```

mm_30.plot()
```

```
#retornos
ret_btc = btc_df['close'].pct_change()
```

```
#Histograma dos Retornos

ret_btc.plot.hist(bins=60);
```

```
#Rolling-Vol - 30 days

vol_30 = ret_btc.rolling(30).std()
```

```
vol_30.plot();
```

```
# Piores quedas dos últimos 30 dias

drawdown_30 = ret_btc.rolling(30).min()
drawdown_30.plot();
```
